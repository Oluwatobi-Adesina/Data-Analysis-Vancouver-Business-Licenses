```python
import pandas as pd

# Example of specifying a local file path
file_path_2023 = '/content/Business Licenses_2023.xlsx'
file_path_2024 = '/content/Business Licenses 2024.xlsx'

# Load the Excel files
df_2023 = pd.read_excel(file_path_2023, engine='openpyxl')
df_2024 = pd.read_excel(file_path_2024, engine='openpyxl')

# Preview the first few rows of each dataset
print("2023 DataFrame")
print(df_2023.head())

print("2024 DataFrame")
print(df_2024.head())
```

```python
# Check for missing values in the 2023 dataset
print("Missing values in the 2023 dataset:")
print(df_2023.isnull().sum())

# Check for missing values in the 2024 dataset
print("Missing values in the 2024 dataset:")
print(df_2024.isnull().sum())
```

```python
# Check the column names in the dataset
print(df_2023.columns)
print(df_2024.columns)
```

```python
# Convert IssueDate and ExpiryDate columns to datetime format for both datasets
df_2023['IssuedDate'] = pd.to_datetime(df_2023['IssuedDate'], errors='coerce')
df_2023['ExpiredDate'] = pd.to_datetime(df_2023['ExpiredDate'], errors='coerce')

df_2024['IssuedDate'] = pd.to_datetime(df_2024['IssuedDate'], errors='coerce')
df_2024['ExpiredDate'] = pd.to_datetime(df_2024['ExpiredDate'], errors='coerce')
```

```python
# Drop rows with missing values in critical columns (e.g., IssueDate or BusinessName)
df_2023.dropna(subset=['IssuedDate', 'BusinessName'], inplace=True)
df_2024.dropna(subset=['IssuedDate', 'BusinessName'], inplace=True)
```

```python
# Fill missing values for non-critical fields (e.g., BusinessType, LicenseType)
df_2023.fillna('Unknown', inplace=True)
df_2024.fillna('Unknown', inplace=True)
```

```python
# Check data types in the 2023 dataset
print(df_2023.dtypes)

# Check data types in the 2024 dataset
print(df_2024.dtypes)
```

```python
# Check the column names in the dataset
print(df_2023.columns)
print(df_2024.columns)

```

```python
import pandas as pd

# The df_2023 and df_2024 are already loaded and cleaned DataFrames

# Concatenate the two datasets for a combined analysis
df = pd.concat([df_2023, df_2024])

# 1. Total number of licenses issued per year (2023 and 2024)
total_licenses_per_year = df.groupby('FOLDERYEAR')['LicenceNumber'].nunique()

# 2. Average and median license fee paid
average_fee = df['FeePaid'].mean()
median_fee = df['FeePaid'].median()

# 3. Distribution of business types
business_type_distribution = df['BusinessType'].value_counts()

# 4. Distribution of licenses by regions (LocalArea)
region_distribution = df['LocalArea'].value_counts()

# 5. Number of employees across businesses
employees_summary = df['NumberofEmployees'].describe()

# 6. Breakdown of license status (active, expired, inactive)
license_status_distribution = df['Status'].value_counts()

# Print results
print("Total number of licenses issued per year:")
print(total_licenses_per_year)

print("\nAverage license fee paid:")
print(average_fee)

print("\nMedian license fee paid:")
print(median_fee)

print("\nDistribution of business types:")
print(business_type_distribution)

print("\nDistribution of licenses by region:")
print(region_distribution)

print("\nSummary of the number of employees across businesses:")
print(employees_summary)

print("\nBreakdown of license status (active, expired, inactive):")
print(license_status_distribution)
```

```python
import pandas as pd

# Define survival based on Status
def is_surviving(status):
    if status == 'Issued':
        return 1  # Surviving
    else:
        return 0  # Not surviving

# Apply the function to the datasets
df_2023['Survival'] = df_2023['Status'].apply(is_surviving)
df_2024['Survival'] = df_2024['Status'].apply(is_surviving)

# Calculate survival rates by year
survival_2023 = df_2023['Survival'].mean() * 100  # Percentage of surviving businesses
survival_2024 = df_2024['Survival'].mean() * 100

print(f"Survival Rate in 2023: {survival_2023:.2f}%")
print(f"Survival Rate in 2024: {survival_2024:.2f}%")

# Calculate survival rates by Business Type for 2023
business_type_survival_2023 = df_2023.groupby('BusinessType')['Survival'].mean() * 100
print("\nSurvival Rate by Business Type in 2023:")
print(business_type_survival_2023)

# Calculate survival rates by Business Type for 2024
business_type_survival_2024 = df_2024.groupby('BusinessType')['Survival'].mean() * 100
print("\nSurvival Rate by Business Type in 2024:")
print(business_type_survival_2024)

# Calculate survival rates by Local Area for 2023
local_area_survival_2023 = df_2023.groupby('LocalArea')['Survival'].mean() * 100
print("\nSurvival Rate by Local Area in 2023:")
print(local_area_survival_2023)

# Calculate survival rates by Local Area for 2024
local_area_survival_2024 = df_2024.groupby('LocalArea')['Survival'].mean() * 100
print("\nSurvival Rate by Local Area in 2024:")
print(local_area_survival_2024)
```
